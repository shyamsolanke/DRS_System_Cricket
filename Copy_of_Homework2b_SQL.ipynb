# Table Creation

Problem 7:

Now we'll create our own table in our database. Separate the Salaries table by Year, and add it back to the database. (You may use basic python to complete the task, however, should still use SQL to query the data.)


```python
listA = ['2011','2012','2013','2014']
print(listA)
for y in listA:
    print(y)
    query = f"CREATE TABLE Table{y} AS SELECT * FROM SALARIES WHERE YEAR='{y}'" # removed SET NOCOUNT ON; from the query
    try:
      df = pd.read_sql(query, conn)
    except Exception as e:
      print(e)
      pass
    df.to_sql(name='Y'+y, con=conn,index="False", if_exists='replace')
```

Problem 10:

What is the average budget used for the top 10 grossing movies?


```python
query = '''SELECT AVG(revenue) AS avgtop10revenue FROM (SELECT revenue, row_number() OVER (ORDER BY revenue DESC) AS rn FROM movies) movies WHERE rn <= 10;'''
df = pd.read_sql(query, conn)
df.head(5)
```

Problem 11:

Which directors have the highest overall voting average? - show the top 5 directors' name and their average rating

```python
query = '''SELECT movies.vote_average as voting_average, directors.name as director_name FROM MOVIES LEFT JOIN DIRECTORS
ON MOVIES.DiRECTOR_ID=DIRECTORS.ID order by voting_average desc LIMIT 5'''
df = pd.read_sql(query, conn)
df.head(5)
```

Problem 12:

What are the top five directors by how much their average budget is?

```python
query = '''SELECT directors.name as director_name, AVG(budget) as average_budget FROM MOVIES LEFT JOIN DIRECTORS
ON MOVIES.DiRECTOR_ID=DIRECTORS.ID group by director_name order by average_budget desc LIMIT 5'''
df = pd.read_sql(query, conn)
df.head(5)
```
